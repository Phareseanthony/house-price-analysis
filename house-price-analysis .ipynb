```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler




data = pd.read_csv("/content/house-price.csv")
data
data.head(50)
#price	area	bedrooms	bathrooms	stories	mainroad	guestroom	basement	hotwaterheating	airconditioning	parking	prefarea furnishingstatus
data_two = data.head(50)
data_two

scaler = StandardScaler()
data[['price1',	'area',	'bedrooms',	"bathrooms",	"stories"]] = scaler.fit_transform (data[['price',	'area',	'bedrooms',	"bathrooms",	"stories"]])
data
label =  LabelEncoder()
data['mainroad'] = label.fit_transform(data['mainroad'])
data= data.sample(200, random_state=46)
data['mainroad']
import matplotlib.pyplot as plt

plt.figure(figsize = (64,48 ))
plt.scatter(data["area"], data["price1"],alpha=0.6)
plt.xlabel("area(standardize)")
plt.ylabel("price1(standardize)")
plt.show()


























```
